```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('LaLiga_Matches_1995-2021.csv')
```

```python
data
```

```python
data_q1 = data
```

```python
data_q1 = data_q1.drop(['Date', 'HTHG' ,'HTAG' , 'HTR' ],axis = 1 )
```

```python
data_q1
```

```python
season2021 = data_q1[data_q1['Season'] == '2020-21']
```

```python
season2021
```

```python
s1 = season2021
s2 = season2021.rename(columns={'HomeTeam': 'AwayTeam', 'AwayTeam': 'HomeTeam'})
```

```python
s2
```

```python
s2['FTR']=s2.apply(lambda x: x['FTR'] if x['FTR'] == 'D' else 'A' if x['FTR'] == 'H' else 'H' ,axis=1)
```

```python
s2
```

```python
full_season = s1.append(s2)
```

```python
full_season
```

```python
full_season['POINT']=full_season.apply(lambda x: 1 if x['FTR'] == 'D' else 3 if x['FTR'] == 'H' else 0 ,axis=1)
```

```python
full_season
```

```python
full_season.groupby(['HomeTeam'])['POINT'].agg([ 'sum']).sort_values('sum',ascending = False)
```
